```python
import requests
import pandas as pd
from io import BytesIO
# Define a URL do arquivo Excel
url = "https://www.gov.br/anp/pt-br/assuntos/precos-e-defesa-da-concorrencia/precos/ppidp/precos-medios-ponderados-semanais-2013.xls"
# Faz uma solicitação HTTP para obter o conteúdo do arquivo Excel
response = requests.get(url)
# Lê o conteúdo do arquivo Excel usando o pandas
df = pd.read_excel(BytesIO(response.content), header=[0,1], skiprows=7)
df.head()
```
